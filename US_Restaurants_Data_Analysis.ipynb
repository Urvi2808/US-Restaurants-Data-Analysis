```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('restaurants.csv')
data.head(5)
```

```python
data.describe()
```

```python
data.drop_duplicates(["id"])
```

```python
data.isnull().sum()
```

```python
data["category"].unique()
```

```python
# Sort the data by ratings in descending order and select the top 10
top_10_restaurants = data.sort_values(by='ratings', ascending=False).head(10)

# Display the names of the top 10 restaurants
print(top_10_restaurants['name'])
```

```python
outlets = data["name"].value_counts()
outlets
```

```python
chains = outlets[outlets >= 2]
single = outlets[outlets == 1]
```

```python
chains
```

```python
print("Total Restaurants = ", data.shape[0])
print("Total Restaurants that are part of some chain = ", data.shape[0] - single.shape[0])
print("Percentage of Restaurants that are part of a chain = ", np.round((data.shape[0] - single.shape[0]) / data.shape[0],2)*100, "%")
```

```python
top10_chains = data["name"].value_counts()[:10].sort_values(ascending=False)
```

```python
sns.barplot(x =top10_chains.values, y =top10_chains.index, palette ='plasma')
# Set labels for x and y axes
plt.xlabel('Number of outlets in US')
plt.ylabel('Name of Restaurant')

# Show the plot
plt.show()
```

```python
outlets = data['name'].value_counts()
```

```python
atleast_5_outlets = outlets[outlets > 4]
```

```python
top10_chains2 = data[data["name"].isin(atleast_5_outlets.index)].groupby("name").mean()["ratings"].sort_values(ascending=False)[:10].sort_values(ascending=False)
```

```python
sns.barplot(x=top10_chains2.values,y = top10_chains2.index, palette='viridis')
# Set labels for x and y axes
plt.xlabel('Number of outlets in US')
plt.ylabel('Name of Restaurant')

# Show the plot
plt.show()
```

```python
cat_count = data.groupby("category").count()["id"].sort_values(ascending=False)[:5]
sns.barplot(x=cat_count.index, y=cat_count.values, palette="muted")

plt.xlabel("Top 5 Category types")
plt.ylabel("Count")
plt.title("Number of restaurants (by Category type)")
plt.xticks(rotation=45, fontsize=7, ha='right')
# Displaying values on top of bars
for index, value in enumerate(cat_count):
    plt.text(index, value + 100, str(value), ha='center', va='bottom')
plt.show()
```
